## WORLD HAPPINESS INDEX FOR YEAR 2022

- Let Analyse these data using Pandas, Numpy, Matplotlib, Seaborn 


```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
happy = pd.read_csv('world-happiness-report-2021.csv')
```

```python
happy.head()
```

```python
happy.shape
```

```python
happy.info()
```

```python
# lets take only required columns
happy_columns = ['Country name','Regional indicator','Ladder score','Logged GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']
```

```python
happy= happy[happy_columns].copy()
```

```python
happy2 = happy.rename({'Country name':'country_name','Ladder score':'ladder_score','Regional indicator': 'regional_indicator','Logged GDP per capita':'logged_GDP_per_capita','Social support':'social_support','Healthy life expectancy':'healthy_life_expectancy','Freedom to make life choices':'freedom_to_make_life_choices','Generosity':'generosity', 'Perceptions of corruption':'perceptions_of_corruption'},axis=1)
```

```python
happy2.head()
```

```python
happy2.isnull().sum()
```

### Plot between Happiness and GDP

```python
sns.set_style('darkgrid')
plt.rcParams['font.size'] =15
plt.rcParams['figure.figsize'] = (10,7)
plt.rcParams['figure.facecolor']='grey'
```

```python
plt.rcParams['figure.figsize'] = (15,7)
plt.title('Plot btw Happiness & GDP')
sns.scatterplot(x=happy2.ladder_score, y=happy2.logged_GDP_per_capita,hue=happy2.regional_indicator)
plt.legend(loc='upper left', fontsize ='10')
plt.xlabel('ladder score')
plt.ylabel('GDP per capita')
```

```python
gdp_region = happy2.groupby('regional_indicator')['logged_GDP_per_capita'].sum()
```

```python
gdp_region
```

```python
gdp_region.plot.pie(autopct= '%1.1f%%')
plt.title('GDP by region')
plt.ylabel('')
```

```python
#total countries 
total_countries =happy2.groupby('regional_indicator')[['country_name']].count()
total_countries
```

```python
##corelation maps#
cor = happy2.corr(method='pearson')
f,ax=plt.subplots(figsize=(10,5))
sns.heatmap(cor,mask=np.zeros_like(cor,dtype=np.bool_),
            cmap='Blues',square=True, ax=ax)
```

```python
##corruption #
corruption = happy2.groupby('regional_indicator')[['perceptions_of_corruption']].mean()
corruption
```

```python
plt.rcParams['figure.figsize'] = (12,8)
plt.title('perceptions_of_corruption in different region')
plt.xlabel('Region', fontsize=15)
plt.ylabel('Corruption Index', fontsize = 15)
plt.xticks(rotation=30, ha='right')
plt.bar(corruption.index,corruption.perceptions_of_corruption)
```

```python
#top and bottom 10 happiest countries
top_10 =happy2.head(10)
bottom_10= happy2.tail(10)
```

```python
fig, axes=plt.subplots(1,2,figsize=(16,6))
plt.tight_layout(pad=2)
xlabels=top_10.country_name
axes[0].set_title('Top 10 happiest countries Life Expectancy')
axes[0].set_xticklabels(xlabels,rotation=45, ha='right')
sns.barplot(x=top_10.country_name,y=top_10.healthy_life_expectancy,ax=axes[0])
axes[0].set_xlabel('Country Name')
axes[0].set_ylabel('Life Expectancy')

xlabels=bottom_10.country_name
axes[1].set_title('Bottom 10 happiest countries Life Expectancy')
axes[1].set_xticklabels(xlabels,rotation=45, ha='right')
sns.barplot(x=bottom_10.country_name,y=bottom_10.healthy_life_expectancy,ax=axes[1])
axes[1].set_xlabel('Country Name')
axes[1].set_ylabel('Life Expectancy')
```

```python
##freedom to make life choices#
plt.rcParams['figure.figsize'] = (15,7)
sns.scatterplot(x=happy2.freedom_to_make_life_choices,y=happy2.ladder_score, hue=happy2.regional_indicator)
plt.legend(loc='upper left', fontsize=10)
plt.xlabel('freedom to make life choices')
plt.ylabel('Happiness Score')
```

```python
#top 10 most corrup countries#
country = happy2.sort_values(by='perceptions_of_corruption').head(10)
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Country with most perception of Corruption')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Corruption index', fontsize=12)
plt.xticks(rotation=30, ha='right')
plt.bar(country.country_name, country.perceptions_of_corruption, color='maroon', edgecolor='black')
```

```python
#bottom 10 perception of corruption#

country = happy2.sort_values(by='perceptions_of_corruption').tail(10)
plt.rcParams['figure.figsize'] = (12,6)
plt.title('Country with most perception of Corruption')
plt.xlabel('Country', fontsize=12)
plt.ylabel('Corruption index', fontsize=12)
plt.xticks(rotation=30, ha='right')
plt.bar(country.country_name, country.perceptions_of_corruption, color='maroon', edgecolor='black')
```

```python
##corruption in terms of happiness score#
plt.rcParams['figure.figsize'] = (15,7)
sns.scatterplot(x=happy2.ladder_score, y=happy2.perceptions_of_corruption, hue=happy2.regional_indicator,)
plt.legend(loc='lower left', fontsize=10)
plt.xlabel('Happiness score')
plt.ylabel('perceptions_of_corruption')
```
